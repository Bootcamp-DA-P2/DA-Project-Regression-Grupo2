```python
import pandas as pd
```

```python
PATH = '../data/raw/competitions.csv'
```

```python
df = pd.read_csv(PATH)
df.head()
```

```python
df = df[['competition_id', 'name']]
df.head()
```

```python
top_5_ligas = ['laliga', 'premier-league', 'serie-a', 'bundesliga']
df = df[df['name'].isin(top_5_ligas)].reset_index(drop=True)
df.head()
```

```python
df.to_csv('../data/clean/competitions_clean.csv', index=False)
```
