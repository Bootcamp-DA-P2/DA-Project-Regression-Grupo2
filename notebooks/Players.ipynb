```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
```

```python
datos_fifa = "../data/raw/players.csv"

df = pd.read_csv(datos_fifa)
df.head()
```

```python
top_4_ligas = ['ES1', 'GB1', 'IT1', 'L1']

df= df[df['current_club_domestic_competition_id'].isin(top_4_ligas)]
```

```python
columnas = [
    'first_name', 'last_name', 'city_of_birth', 'sub_position', 'foot',
    'agent_name', 'image_url', 'international_caps', 'international_goals','url', 'player_code', 'country_of_citizenship', 'highest_market_value_in_eur'
    
]

df = df.drop(columns=columnas)

df.info()
```

```python
df.info()

print(df.describe())
```

```python
df['date_of_birth'] = pd.to_datetime(df['date_of_birth'])

df['contract_expiration_date'] = pd.to_datetime(df['contract_expiration_date'])

df['player_id'] = df['player_id'].astype(int)

df['last_season'] = df['last_season'].astype(int)

df['current_club_id'] = df['current_club_id'].astype(int)

df['market_value_in_eur'] = pd.to_numeric(df['market_value_in_eur'], errors='coerce').fillna(0).astype(int)


# Cambio formato numeros

cols_text = [
    'name', 'position', 'country_of_birth', 'current_club_name'
]
df[cols_text] = df[cols_text].astype(str)
```

```python
print(df.duplicated().sum())

print(df[df.duplicated(keep=False)])
```

```python
print(df.isnull().sum())
```

```python
# Porcentaje de nulos por columna para evaluar si vale la pena conservar o eliminar una variable.
print((df.isnull().sum() / len(df)) * 100)
```

```python
# Eliminación de la columna 'contract_expiration_date' y 'current_national_team_id' por tener un alto porcentaje de valores nulos.
df = df.drop(columns=['contract_expiration_date', 'current_national_team_id'])

#Eliminación de los registros que tienen nulos en 'date_of_birth'
df = df.dropna(subset=['date_of_birth', 'country_of_birth'])
```

```python
print(df.isnull().sum())
```

```python
df = df.fillna(df['height_in_cm'].mean())
```

```python
# Creación de la columna 'age' a partir de 'date_of_birth'
df['age'] = (pd.Timestamp.now() - df['date_of_birth']).dt.days // 365

# Eliminación de la columan 'date_of_birth'
df = df.drop(columns='date_of_birth')
```

```python
df = df[df['age']<=35]
```

```python
len(df)
```

```python
df = df.rename(columns={'current_club_id': 'club_id', 'current_club_domestic_competition_id':'league_id', 'current_club_name':'club_name'})
```

```python
df.to_csv("../data/clean/players_clean.csv", index=False)
```
