```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
```

```python
datos_fifa = "../data/raw/players.csv"

df = pd.read_csv(datos_fifa)
df.head()
```

```python
columnas = [
    'first_name', 'last_name', 'last_season', 'city_of_birth', 'sub_position', 'foot',
    'agent_name', 'image_url', 'international_caps', 'international_goals','url', 'player_code', 'country_of_citizenship'
    
]

df = df.drop(columns=columnas)

df.info()
```

```python
df.info()

print(df.describe())
```

```python
df['date_of_birth'] = pd.to_datetime(df['date_of_birth'])

df['contract_expiration_date'] = pd.to_datetime(df['contract_expiration_date'])

df['player_id'] = df['player_id'].astype(int)

df['current_club_id'] = df['current_club_id'].astype(int)

df['market_value_in_eur'] = pd.to_numeric(df['market_value_in_eur'], errors='coerce').fillna(0).astype(int)

df['highest_market_value_in_eur'] = pd.to_numeric(df['highest_market_value_in_eur'], errors='coerce').fillna(0).astype(int)


# Cambio formato numeros

cols_text = [
    'name', 'position', 'country_of_birth', 'current_club_name'
]
df[cols_text] = df[cols_text].astype(str)
```

```python
print(df.duplicated().sum())

print(df[df.duplicated(keep=False)])
```

```python
print(df.isnull().sum())

# Calcula el porcentaje de nulos por columna para evaluar si vale la pena conservar o eliminar una variable.
print((df.isnull().sum() / len(df)) * 100)

```

```python
df = df.fillna(0)

print(df.isnull().sum())
```

```python
df.to_csv("../data/clean/players_clean.csv", index=False)
```
