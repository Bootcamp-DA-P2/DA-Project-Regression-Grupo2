```python
import pandas as pd
```

```python
PATH = '../data/raw/clubs.csv'
df = pd.read_csv(PATH)
df.head()
```

```python
df = df[['club_id', 'name', 'domestic_competition_id']]
df = df.rename(columns={'domestic_competition_id': 'league_id', 'name':'club_name'})
df.head()
```

```python
ligas_8 = ['ES1', 'GB1', 'IT1', 'L1', 'ARG1', 'AUS1', 'COL1', 'JAP1']
df = df[df['league_id'].isin(ligas_8)].reset_index(drop=True)
df.head()
```

```python
df.shape[0]
```

```python
df.to_csv('../data/clean/clubs_clean.csv', index=False)
```
